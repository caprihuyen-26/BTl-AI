```python
import os
import yaml
from collections import defaultdict

# Đường dẫn tới thư mục chứa train/val/test
dataset_dir = '/content/drive/MyDrive/PPE'

# Đọc file YAML để lấy tên nhãn
data_yaml_path = os.path.join(dataset_dir, 'data.yaml')
with open(data_yaml_path, 'r') as f:
    data = yaml.safe_load(f)

label_names = data['names']
label_counts = defaultdict(int)  # dict để đếm số lần xuất hiện của mỗi class

# Duyệt qua 3 tập: train, val, test
splits = ['train', 'val', 'test']

for split in splits:
    labels_path = os.path.join(dataset_dir, split, 'labels')
    if not os.path.exists(labels_path):
        continue

    for label_file in os.listdir(labels_path):
        if not label_file.endswith('.txt'):
            continue

        with open(os.path.join(labels_path, label_file), 'r') as f:
            for line in f:
                if line.strip() == '':
                    continue
                class_id = int(line.strip().split()[0])
                label_counts[class_id] += 1

# In kết quả
print("Số lượng nhãn xuất hiện : ")
for class_id, count in sorted(label_counts.items()):
    print(f"{class_id}: {label_names[class_id]} :{count}")
```

## Training Hyperparameters


```python
IMAGE_SIZE = 640
BATCH_SIZE=16
EPOCHS=10
```

## Load pretrained model

```python
model = YOLO("/content/drive/MyDrive/PPE/yolov12m.pt")
```

## Fine-tuning model

```python
model.train(
    data="data.yaml",
    epochs=EPOCHS,
    imgsz=IMAGE_SIZE,
    batch=BATCH_SIZE,
    name=f"yolov12m-{EPOCHS}-epochs",
     project="/content/drive/MyDrive/PPE/runs/train",
    save = True,
    save_period = 1,
    val=True

)
```

```python
import time

# Load mô hình đã huấn luyện
model_path = "/content/drive/MyDrive/PPE/runs/train/yolov12m-10-epochs/weights/best.pt"
model = YOLO(model_path)

# Mở video input
video_path = "/content/drive/MyDrive/PPE/input_video1.mp4"
cap = cv2.VideoCapture(video_path)

# Lấy thông tin video
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps_input = cap.get(cv2.CAP_PROP_FPS)

# Tạo video output
output_path = "/content/drive/MyDrive/PPE/output_video1.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps_input, (width, height))

# Danh sách nhãn và màu sắc
class_names = ['Boots', 'Helmet', 'Human', 'Vest']
colors = {
    'Boots':  (0, 255, 255),
    'Helmet': (255, 0,   0),
    'Human':  (0, 255,   0),
    'Vest':   (0,   0, 255)
}

# Thống kê
frame_count = 0
inference_count = 0
inference_time_total = 0
cached_boxes = []
last_fps = 0.0

# Chạy inference mỗi frame
inference_interval = 1

start_time = time.time()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    annotated_frame = frame.copy()

    # Chỉ chạy model mỗi inference_interval frame
    if frame_count % inference_interval == 0:
        t0 = time.time()
        results = model(
            frame,
            conf=0.5,       # giữ threshold confidence model
            iou=0.3,        # giảm threshold IoU cho NMS
            max_det=100,    # tăng số box tối đa
            imgsz=1024,     # resize đầu vào (tùy GPU)
            verbose=False
        )[0]

        inference_time = time.time() - t0
        inference_time_total += inference_time
        inference_count += 1
        last_fps = 1.0 / (inference_time + 1e-8)

        # Cập nhật cached_boxes
        cached_boxes = []
        if results.boxes is not None:
            boxes  = results.boxes.xyxy.cpu().numpy()
            scores = results.boxes.conf.cpu().numpy()
            classes= results.boxes.cls.cpu().numpy().astype(int)

            for box, score, cls in zip(boxes, scores, classes):
                # hạ ngưỡng lọc confidence để giữ cả box nhỏ
                if score >= 0.3:
                    cached_boxes.append((box, score, cls))

    # Vẽ nhãn từ cached_boxes
    for box, score, cls in cached_boxes:
        x1, y1, x2, y2 = map(int, box)
        label = f"{class_names[cls]} {score:.2f}"
        color = colors.get(class_names[cls], (255,255,255))
        cv2.rectangle(annotated_frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(annotated_frame, label, (x1, y1-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # Vẽ FPS inference
    cv2.rectangle(annotated_frame, (10, 10), (170, 50), (0,0,0), -1)
    cv2.putText(annotated_frame, f"FPS inf: {last_fps:.2f}", (10, 35),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)

    out.write(annotated_frame)
    frame_count += 1

# Thống kê và giải phóng
total_time = time.time() - start_time
avg_fps = frame_count / total_time
avg_infer_fps = (inference_count / inference_time_total) if inference_time_total>0 else 0

print(f"Tổng frame: {frame_count}")
print(f"Frame chạy model: {inference_count}")
print(f"Tổng thời gian: {total_time:.2f} giây")
print(f"Tổng thời gian chạy inference: {inference_time_total:.2f} giây")
print(f"FPS trung bình (toàn bộ): {avg_fps:.2f}")
print(f"FPS trung bình (inference): {avg_infer_fps:.2f}")



cap.release()
out.release()
```
